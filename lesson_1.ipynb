````python
import matplotlib.pyplot as plt #importing matplotlib

y = [20,10,5,0] #list for X 
x = [1,15,20,25] #list for Y

plt.plot(x,y)
plt.ylabel("laughs") #labels for the graphs
plt.xlabel("jokes")
plt.title("Why your not funny")
````
