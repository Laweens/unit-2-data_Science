```python
import matplotlib.pyplot as plt 
import pandas as pd 

df = pd.read_csv("Summer_sports_Experience.csv")
df
```

```python
df.groupby(['Borough Location', 'Park Location'])['Attendance Sum'].sum()
```

```python
park_att = df.groupby(['Borough Location', 'Park Location'])['Attendance Sum'].sum()
print()
plt.style.use('dark_background')

park_att.plot(kind='bar')
plt.ylim(1000,75000)

plt.rcParams["figure.facecolor"] = 'white'
plt.rcParams["axes.facecolor"] = 'white'
plt.rcParams["savefig.facecolor"] = 'black'
plt.rcParams['axes.labelcolor'] = 'black'
plt.rcParams['xtick.color'] = 'black'
plt.rcParams['ytick.color'] = 'black'
COLOR = 'black'






```

The graph shows the sum of attendance in parks throughout the Boroughs in NYC with the Queens playgrground for all children having a far higher average attendance sum than any other park in the Boroughs.
