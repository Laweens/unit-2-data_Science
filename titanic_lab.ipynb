```python
import matplotlib.pyplot as plt 
import pandas as pd 

df = pd.read_csv("titanic.csv")
df.head()
sex = df['Sex']
print(sex)
```

```python
male = 0
female = 0 

for i in sex:
    i = i.strip()
    if i == "male":
        male += 1
    elif i =="female":
        female += 1

x = [male,female]
labels = ['Female', 'Male']
y = [male, female]
fig, ax = plt.subplots()
ax.pie(x,labels = labels, autopct= '%.1f%%')
plt.ylabel("Amount", fontsize = 15 ) #increased the font size
plt.xlabel("Gender", fontsize = 25 )
plt.title("Male to female ratio ")
plt.pie(y)
```
